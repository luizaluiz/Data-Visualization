```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5)) #Esta linha cria uma figura mestre contendo uma linha e duas colunas de gráficos, definindo o tamanho total da imagem como 15 por 5 polegadas.

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes']) #No primeiro espaço (à esquerda), o código desenha um gráfico de linhas usando os anos no eixo horizontal e a quantidade de imigrantes no eixo vertical.
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013') #Define o título do primeiro gráfico, inserindo uma quebra de linha para organizar o texto.
axs[0].set_xlabel('Ano') #Adiciona o rótulo "Ano" na parte inferior do primeiro gráfico.
axs[0].set_ylabel('Número de imigrantes') #Adiciona o rótulo indicando a quantidade de pessoas na lateral esquerda do primeiro gráfico.
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5)) #Configura o eixo do tempo para mostrar marcações (números) de 5 em 5 anos, evitando que as datas fiquem amontoadas.
axs[0].grid() #Esta instrução ativa as linhas de grade (o famoso "quadriculado") apenas no primeiro gráfico (o de linha), facilitando a identificação de qual valor o ponto no gráfico atinge no eixo vertical em relação a cada ano.

axs[1].boxplot(dados_brasil['imigrantes']) #No segundo espaço (à direita), o código cria um gráfico de caixa (boxplot) para mostrar a variação, a média e possíveis valores atípicos dos dados.
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013') #Define o título explicativo para o segundo gráfico.
axs[1].set_xlabel('Brasil') #Identifica que os dados do boxplot referem-se especificamente ao Brasil no eixo horizontal.
axs[1].set_ylabel('Número de imigrantes') # Repete o rótulo lateral para manter a padronização da escala de imigrantes no segundo gráfico.
axs[1].grid() #Esta linha ativa as linhas de grade no segundo gráfico (o boxplot), o que é muito útil para alinhar visualmente os limites da "caixa" e os valores das extremidades (os "bigodes") com os números de imigrantes no eixo vertical.

plt.show() #Por fim, esta linha ordena que o computador exiba a janela com os dois gráficos finalizados na tela.
```
